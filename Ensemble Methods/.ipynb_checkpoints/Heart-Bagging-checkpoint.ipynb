```python
upper_limit = df['MaxHR'].mean() + 3*df['MaxHR'].std()
lower_limit = df['MaxHR'].mean() - 3*df['MaxHR'].std()

df['MaxHR'] = np.where(
    df['MaxHR']>upper_limit,
    upper_limit,
    np.where(
        df['MaxHR']<lower_limit,
        lower_limit,
        df['MaxHR']
    )
)
```

```python
sex_le = LabelEncoder()
chest_le = LabelEncoder()
rest_le = LabelEncoder()
excercise_le = LabelEncoder()
st_le = LabelEncoder()
```

```python
df['sex_n'] = sex_le.fit_transform(df['Sex'])
df['chest_n'] = chest_le.fit_transform(df['ChestPainType'])
df['rest_n'] = rest_le.fit_transform(df['RestingECG'])
df['excercise_n'] = excercise_le.fit_transform(df['ExerciseAngina'])
df['st_n'] = st_le.fit_transform(df['ST_Slope'])
```

```python
df.head()
```

```python
df = df.drop(['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'], axis = 'columns')
df.head()
```

```python
X = df.drop(['HeartDisease'], axis = 'columns')
Y = df.HeartDisease
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, x_test, Y_train, y_test = train_test_split(X_scaled, Y, test_size = 0.3)
```
